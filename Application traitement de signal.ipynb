# Exercice
## Dans cet exercice nous utilisons une image importée à partir du module 'misc' de la bibliothèque 'scipy'.
## Le script est le suivant:

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import misc 
image=plt.imread('bacterie_0.jpg')
image=image[:,:,0]
plt.imshow(image)
plt.show()
```

```python
image.shape
```

### 1) Ajouter un bruit aléatoire à l'image et afficher le résultat

```python

noise_image=image+3*image.std()*np.random.randn(image.shape[0],image.shape[1])
image=np.uint8((noise_image/noise_image.max())*255)
```

```python
plt.imshow(image,cmap='gray')
```

### 2) Comparaison des deux histogrammes

```python
plt.figure(figsize=(12,8))
plt.hist(image.ravel(),bins=20,color='blue',edgecolor='black',alpha=0.5,density='True')
plt.hist(image.ravel(),bins=20,color='red',edgecolor='black',alpha=0.3,density='True')
plt.show()
```

# Exercice 2

```python
image=plt.imread('bacterie_0.jpg')
image=image[:,:,0]
plt.imshow(image,cmap='gray')
plt.show()
```

```python
image_2=np.copy(image)
plt.hist(image_2.ravel(),bins=255)
plt.show()
```

```python
#On isole maintenant les bactéries
image=image>20
plt.imshow(image)
plt.show()
```

```python
# On fait la morphologie pour éliminer les artéfacts
from scipy import ndimage
open_x=ndimage.binary_opening(image)
plt.imshow(open_x)
```

```python
labels_image,n_labels=ndimage.label(open_x)
print(n_labels)
```

```python
plt.imshow(labels_image)
plt.colorbar()
```

```python
sizes=ndimage.sum(open_x,labels_image,range(n_labels))
sizes
# Comment utiliser le sizes sup 40 pour prendre juste les cellules qui respectent 
#cette dimension
```

```python
plt.scatter(range(n_labels), sizes, c='blue',s=100)
```

```python
sizes_norm=[]
for i in range(n_labels):
    if sizes[i,]>230:
        sizes_norm.append(sizes[i,])
print(sizes_norm)
```

```python
n=len(sizes_norm)
n
```

```python
plt.scatter(range(n),sizes_norm,c=sizes_norm,s=100)
plt.colorbar()
plt.show()
```

### Calculons la taille moyenne, l'écart-type et la variance des bactéries

```python
moy=np.mean(sizes_norm)
```

```python
var=np.var(sizes_norm)
```

```python
np.std(sizes_norm)
```

```python
print('La taille moyenne des bactéries est égale à {} La dispersion autour de cette taille moyenne est de {}'.format(moy,var))
```
